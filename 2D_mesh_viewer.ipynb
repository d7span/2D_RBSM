```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ファイルの読み取り
f = open('ELEMENT.INDAT', 'r', encoding = "utf-8_sig")
text = f.readline() #要素数 最大頂点数
text = f.readline() 
text = text.strip()
text = text.split()
NELM = int(text[0])
MAXNODE = int(text[1])


#配列の作成&初期化
TEN = np.zeros(NELM,dtype=np.int64)
NODE = np.zeros([NELM,MAXNODE],dtype=np.int64)

text = f.readline() #頂点数, 節点番号
for i in range(0,NELM):
    text = f.readline()     
    text = text.strip()
    text = text.split()
    TEN[i] = int(text[1])
    for j in range(0,TEN[i]):
        NODE[i][j] = int(text[j+2])
    #print(i,TEN[i], NODE[i][0],NODE[i][1])

text = f.readline() #節点数
text = f.readline()
text = text.strip()
NUMNODE = int(text)


#配列の作成&初期化
XNODE = np.zeros(NUMNODE,dtype=np.float64)
YNODE = np.zeros(NUMNODE,dtype=np.float64)

text = f.readline() #X座標, #Y座標
for i in range(0,NUMNODE):
    text = f.readline() 
    text = text.strip() 
    text = text.split()
    XNODE[i] = text[1]
    YNODE[i] = text[2]
    #print(XNODE[i],YNODE[i])
    
print("NELM",NELM,"MAXNODE",MAXNODE,"NUMNODE",NUMNODE)
```

```python
fig, ax = plt.subplots(figsize=(16, 12.0))
ax.set_aspect("equal")
color=(0.85,0.85,0.85,0.9)

#temporaryなリスト作成
_x = [] 
_y = []
for i in range(0,NELM):
    for j in range(0,TEN[i]):
        node = NODE[i][j]
        _x.append(XNODE[node-1])
        _y.append(YNODE[node-1])       
    ax.fill(_x,_y,edgecolor="black",facecolor="white")
    _x.clear()
    _y.clear()
```
